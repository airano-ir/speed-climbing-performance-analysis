### Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø±Ù‡Ø§ÛŒ Ù…Ù‚Ø§ÛŒØ³Ù‡â€ŒØ§ÛŒ

```python
# Ù…Ù‚Ø§ÛŒØ³Ù‡ Ù…Ø³ØªÙ‚ÛŒÙ… vertical trajectory

fig, ax = plt.subplots(figsize=(14, 6))

# Left climber
left_df['com_y_flipped'] = 1 - left_df['kp_23_y']
left_df['time_norm'] = left_df['timestamp'] - left_df['timestamp'].min()
ax.plot(left_df['time_norm'], left_df['com_y_flipped'], 
        linewidth=2.5, color='blue', label='Left Climber', alpha=0.8)

# Right climber
right_df['com_y_flipped'] = 1 - right_df['kp_23_y']
right_df['time_norm'] = right_df['timestamp'] - right_df['timestamp'].min()
ax.plot(right_df['time_norm'], right_df['com_y_flipped'], 
        linewidth=2.5, color='red', label='Right Climber', alpha=0.8)

ax.set_xlabel('Time (seconds)', fontsize=13, fontweight='bold')
ax.set_ylabel('Vertical Position (normalized)', fontsize=13, fontweight='bold')
ax.set_title('Vertical COM Trajectory - Comparison', fontsize=15, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=12)

plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/comparison_vertical.png", dpi=300, bbox_inches='tight')
plt.show()

print("âœ“ Comparison plot saved!")
```
